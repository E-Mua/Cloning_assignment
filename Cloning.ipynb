```python
import pandas as pd
import seaborn as sns

# Load Titanic dataset
df = sns.load_dataset('titanic')
print(df.head())
df.columns
```

```python
import matplotlib.pyplot as plt

# Count values for the 'sex' column
sex_counts = df['sex'].value_counts()

# Plot pie chart
plt.figure(figsize=(7, 7))
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightcoral'])
plt.title('Distribution of Gender in Titanic')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.bar(df['class'], df['fare'])
plt.xlabel("class")
plt.ylabel("fare")
plt.title("FARE PER CLASS")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['fare'].dropna(), bins=30, kde=True, color="blue")
plt.title("Titanic Passenger Fare Distribution")
plt.xlabel("fare")
plt.ylabel("Count")
plt.show()
```
